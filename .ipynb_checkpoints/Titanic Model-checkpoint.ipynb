# 3.2.2 Numerical Variable

SibSp, Parch, Age and Fare

```python
def detect_outliers(df, n, features):
    outlier_indices = []
    for col in features:
        Q1 = np.percentile(df[col], 25)
        Q3 = np.percentile(df[col], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = [key for key, value in outlier_indices.items() if value > n]
    return multiple_outliers

outliers_to_drop = detect_outliers(train, 2, ['Age', 'SibSp', 'Parch', 'Fare'])
print("We will drop these {} indices: ".format(len(outliers_to_drop)), outliers_to_drop)
```

```python
# Outliers in numerical variable
train.loc[outliers_to_drop, : ]
```

```python
# Drop outliers and reset index
print("Before: {} rows".format(len(train)))
train = train.drop(outliers_to_drop, axis=0).reset_index(drop=True)
print("After: {} rows".format(len(train)))
```

# 4. Data Preprocessing 

# 4.1 Drop and fill missing values

```python
# Drop ticket and cabin fetures from training and test set
train = train.drop(['Ticket', 'Cabin'], axis=1)
test = test.drop(['Ticket', 'Cabin'], axis=1)

```

```python
test.head()
```

```python
#Missing values in training set
train.isnull().sum().sort_values(ascending=False)
```

```python
# Compute the most frequent value of Embarked in training set
mode = train['Embarked'].dropna().mode()[0]
mode
```

```python
# Fill missing value in Embarked with mode
train['Embarked'].fillna(mode, inplace = True)
```

```python
#Missing values in test set
test.isnull().sum().sort_values(ascending=False)
```

```python
# Compute median of Fare in test set
median = test['Fare'].dropna().median()
median
```

```python
# Fill missing value in Fare with median
test['Fare'].fillna(median, inplace = True)
```

```python
# Combine training set and test set
combine = pd.concat([train, test], axis = 0).reset_index(drop = True)
combine.head()
```

```python
#Missing values in combined dataset
combine.isnull().sum().sort_values(ascending=False)
```

```python
# Convert Sex into numerical values where 0 = male and 1 = female
combine['Sex'] = combine['Sex'].map({'male': 0, 'female': 1})
```

```python
sns.catplot(y='Age', x='Sex', hue='Pclass', kind='box', data=combine)
sns.catplot(y='Age', x='Parch', kind='box', data=combine)
sns.catplot(y='Age', x='SibSp', kind='box', data=combine) 
```

```python
#sns.heatmap(combine.drop(['Survived','Name','PassengerId','Fare'], axis = 1).corr(), annot = True, cmap ='coolwarm')
sns.heatmap(combine.drop(['Survived','Name','PassengerId','Fare'], axis = 1).corr(), annot = True, cmap ='coolwarm')
```

```python
# Check number of missing ages
#age_nan_indices = list(combine[combine['Age'].isnull()].index)
#len(age_nan_indices)
age_nan_indices = list(combine[combine['Age'].isnull()].index)
len(age_nan_indices)
```

```python
# loop through list and input missing Ages
for index in age_nan_indices:
    median_age = combine['Age'].median()
    predict_age = combine['Age'][(combine['SibSp'] == combine.iloc[index]['SibSp'])
                                 & (combine['Parch'] == combine.iloc[index]['Parch'])
                                 & (combine['Pclass'] == combine.iloc[index]['Pclass'])].median()
    if np.isnan(predict_age):
        combine['Age'].iloc[index] = median_age
    else:
        combine['Age'].iloc[index] = predict_age
```

```python
# Make sure there is no more missing ages
combine['Age'].isnull().sum()
```

```python
# passenger fare distribution
sns.distplot(combine['Fare'], label='Skewness: %.2f'%(combine['Fare'].skew()))
plt.legend(loc='best')
plt.title('Passenger Fare Distribution')
```

```python
# Apply log transformation to fare column to reduce skewness
combine['Fare'] = combine['Fare'].map(lambda x: np.log(x) if x > 0 else 0)
```

```python
# Passenger fare distribution after log transformation
sns.distplot(combine['Fare'], label = 'Skewness: %.2f'%(combine['Fare'].skew()))
plt.legend(loc = 'best')
plt.title('Passenger Fare Distribution After Log Transformation')
```

# 5.3 Training Accuracy

```python
models = pd.DataFrame({'Mdel' : ['Support Vector Machines','Logistic Regression', 'K-Nearest Neighours',
                                'Random Forest', 'Linear SVC','Decision Tree','Linear Regration'],
                      'Score' : [acc_svc,acc_log,acc_knn,acc_random_forest,acc_linear_svc,acc_decision_tree,
                                acc_linear_reg]})
models.sort_values(by = 'Score', ascending = False, ignore_index = True)
```

# 5.3.1 K-Fold crossvalidation

```python
# crate a list which contains classifier
classifiers = []
classifiers.append(LogisticRegression())
classifiers.append(SVC())
classifiers.append(KNeighborsClassifier(n_neighbors=5))
classifiers.append(DecisionTreeClassifier())
classifiers.append(LinearSVC())
classifiers.append(LinearRegression())
classifiers.append(RandomForestClassifier())

len(classifiers)
```

```python
# Perform cross-validation for each classifier
cv_results = []
for classifier in classifiers:
    cv_results.append(cross_val_score(classifier, x_train, y_train, scoring='accuracy', cv=7))
```

```python
cv_mean = []
cv_std = []
for cv_result in cv_results:
    cv_mean.append(cv_result.mean())
    cv_std.append(cv_result.std())
```

```python
cv_res = pd.DataFrame({'Cross validation Mean' : cv_mean, 'Cross Validation Std': cv_std, 'Algorithm':['Logistic Regression','Support Vector Machines','K-Nearest Neighours','Random Forest', 'Linear SVC','Decision Tree','Linear Regration']})
cv_res.sort_values(by = 'Cross validation Mean', ascending=False, ignore_index=True)
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
grid.fit(x_train, y_train)
```

```python
print("Best Parameter: ", grid.best_params_)
print("Best estimator: ", grid.best_estimator_)
```

```python
# Training accuracy
svc = SVC(C=100, gamma = 0.01, kernel = 'rbf')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
acc_svc = round(svc.score(x_train, y_train) * 100, 2)

print("Accuracy:", acc_svc)
```

```python
#Mean cross validation score
cross_val_score(svc,x_train, y_train, scoring='accuracy', cv=7).mean()
```
